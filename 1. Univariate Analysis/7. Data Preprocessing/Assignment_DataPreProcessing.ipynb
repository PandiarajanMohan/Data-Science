```python
#Check for column names and find error there itself!!
dataset.columns=dataset.columns.str.strip() # Trim the space from the column names
#dataset['Discounts']=dataset['Discounts'].str.strip()
```

```python
df=dataset.copy()
```

# Remove '$' sign and '-' from all columns where they are present
# Change datatype from objects to int after the above two.
# Challenge: Try removing " , " (comma) from all numerical numbers.

```python
quan=['Units Sold','Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts','Sales', 'COGS', 'Profit']
for col in quan:
    df[col] = df[col].astype(str)         
    df[col] = df[col].str.replace(r'[^0-9.-]', '', regex=True) 
    df[col] = df[col].replace({'':'0','-': '0'}) 
    df[col]=df[col].str.strip()
    df[col] = df[col].astype(float) 
```
